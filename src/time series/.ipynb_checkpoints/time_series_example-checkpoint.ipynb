### Snap periods data: Detection of the cannibalisation and halo effects
Calculate Halo and Cannibalisation effects on the Walmart (M5) data.
- This notebook extracts the sales per every product within the department and also the promotions, as per the events described in the calendar.
- Then it calculates the daily contribution of each item to the total department sales.
- It also flags the availabily of the product based on the total sales of the stores. If zero, the product is considered unavailable (ie: seasonal products).
- Then the average quota/contribution of every SKU is calculated splitting by promotional/non-promotional taking into account if the product was available.
- Then the ratio of promo/non-promo is a square matrix (size prods x prods) where each column represents the item on promotion and its effects of the other items.
- We infer cannibalisation/haloing as per the threshold $\eta$:
    - Cannibalisation from product A: ratio of the  B changes $\lt 1-\eta$ when A on promotion
    - Haloing from product A: ratio of the product B changes $\gt\eta$ when A on promotion
    - Still working out $\eta$, at the moment $\eta$=0.25
- Add to this explanation that the contributions reflect that the category sales are stable if no cannibalisation/haloing occurs.

Cannibalisation means that a product _A_ that increases its sales with respect to its baseline sales it is also stealing sales from other intra-category products, reflected in the category contribution.

```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fcn_helpers as fhelp
import fcn_simulation as fsim
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split

from sktime.datasets import load_airline
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.performance_metrics.forecasting import smape_loss
from sktime.utils.plotting.forecasting import plot_ys

#plt.rcParams['figure.figsize'] = [14, 7]
fhelp.makeFolder('results')
pd.options.display.max_rows = None
# Use v2.0 colour cycle
def_colours = plt.rcParams['axes.prop_cycle'].by_key()['color']
# Fig sizes
fig_h = 10
fig_w = 18
```

### Read the category data from the files

```python
outputBaseFolder = '~/Google Drive/order/Machine Learning Part/data/Walmart(M5)'
dept_id = 'FOODS_1'
glob_pattern = os.path.join(os.path.expanduser(outputBaseFolder), '*', dept_id, '*', 'national', '*.pickle')
national_level_files = glob.glob(glob_pattern)
max_stores = 10
```

For all of the products within the category, get the baseline sales and the events indicator. Also calculate the category's total sales and total volume. 
Some academics argue that the volume should not fluctuate much.

```python
flag_promos_raw_data = 'total_snapstores'
threshold = 0.35

all_dfs_snap = []
all_dfs_regular = []
all_dfs_snap_flag = []
all_dfs_sales = []

all_dfs_snap_norm = []
all_dfs_regular_norm = []


for this_file in national_level_files:
    df_A = fhelp.readPickleFile(this_file)    

    item_id = df_A.item_id[0]
    
    var_base_sales = 'baseline_sales_' + item_id
    var_snap_sales = 'snap_sales_' + item_id
    var_promos = 'snap_stores_' + item_id
    var_total_sales = 'total_sales' + item_id
    
    df_A.rename(columns={'total_baseline_sales': var_base_sales,
        'total_snapsales': var_snap_sales,
        flag_promos_raw_data: var_promos,
        'total_sales': var_total_sales}, inplace=True)
    
    # Sales normalised per store
    snap_sales_norm = df_A[var_snap_sales].copy()
    idx_normalise = df_A[var_promos] > 0
    snap_sales_norm[idx_normalise] /= df_A[var_promos][idx_normalise]
    all_dfs_snap_norm.append(snap_sales_norm)

    reg_sales_norm = df_A[var_base_sales].copy()
    idx_normalise = df_A['total_baseline_stores'] > 0
    reg_sales_norm[idx_normalise] /= df_A['total_baseline_stores'][idx_normalise]
    all_dfs_regular_norm.append(reg_sales_norm)

    all_dfs_regular.append(df_A[var_base_sales])
    all_dfs_snap.append(df_A[var_snap_sales])
    all_dfs_snap_flag.append(df_A[var_promos])
    all_dfs_sales.append(df_A[var_total_sales])
    
df_regular = pd.concat(all_dfs_regular, axis=1)
df_snap = pd.concat(all_dfs_snap, axis=1)
df_snap_flag = pd.concat(all_dfs_snap_flag, axis=1)

# total sales per product
df_total_sales = pd.concat(all_dfs_sales, axis=1)

# sales normalised per store 
df_regular_norm = pd.concat(all_dfs_regular_norm, axis=1)
df_snap_norm = pd.concat(all_dfs_snap_norm, axis=1)

# Daily quota w.r.t. category/department sales
# Snap
category_total_snap_sales = df_snap.sum(axis=1).values
snap_total_items = (df_snap>0).sum(axis=1).astype(float).values
category_total_snap_sales_norm = category_total_snap_sales.copy()
idx_valid = snap_total_items > 0
category_total_snap_sales_norm[idx_valid] /= snap_total_items[idx_valid]

# Regular
category_total_regular_sales = df_regular.sum(axis=1).values
regular_total_items = (df_regular>0).sum(axis=1).astype(float).values
category_total_regular_sales_norm = category_total_regular_sales.copy()
idx_valid = regular_total_items > 0
category_total_regular_sales_norm[idx_valid] /= regular_total_items[idx_valid]


category_total_sales = df_total_sales.sum(axis=1).values

# Account for products being seasonal or not available (total sales > 0)
df_sku_availability = df_total_sales > 0
```

#### Plot the units vs dates

```python
x_axis = df_A.date
y_axis = category_total_sales
f, ax = plt.subplots(figsize=(fig_w, fig_h/2))
ax.plot(x_axis, y_axis, label='All sales', color=def_colours[0], linewidth=2)
ax.plot(x_axis,df_snap.sum(axis=1), label='Snap sales', color=def_colours[1], alpha=0.65)
#ax.plot(x_axis,df_regular.sum(axis=1), label='Regular sales')
plt.xlabel('dates')
plt.ylabel(f'Department unit sales for {dept_id}')
plt.grid(True)
plt.legend()
plt.show()
```

# Evaluate sktime

```python
y_train, y_test = temporal_train_test_split(df_total_sales.sum(axis=1), test_size=90)
plot_ys(y_train, y_test, labels=["y_train", "y_test"])
print(y_train.shape[0], y_test.shape[0])
```

```python
reg_period = 7
fh = np.arange(len(y_test)) + 1
from sktime.forecasting.naive import NaiveForecaster
forecaster = NaiveForecaster(strategy="seasonal_last", sp=reg_period)
forecaster.fit(y_train)
y_pred = forecaster.predict(fh)
plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"])
smape_loss(y_pred, y_test)
```

```python
from sktime.forecasting.compose import ReducedRegressionForecaster
from sklearn.neighbors import KNeighborsRegressor

reg_period = 28

regressor = KNeighborsRegressor(n_neighbors=2)
forecaster = ReducedRegressionForecaster(regressor=regressor, window_length=reg_period, strategy="recursive")
forecaster.fit(y_train)
y_pred = forecaster.predict(fh)
plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"]);
smape_loss(y_test, y_pred)
```

```python
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
forecaster = ExponentialSmoothing(trend="add", seasonal="multiplicative", sp=reg_period)
forecaster.fit(y_train)
y_pred = forecaster.predict(fh)
plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"]);
smape_loss(y_test, y_pred)
```

```python
from sktime.forecasting.compose import EnsembleForecaster
forecaster = EnsembleForecaster([
    ("ses", ExponentialSmoothing(seasonal="multiplicative", sp=reg_period)),
    ("holt", ExponentialSmoothing(trend="add", damped=False, seasonal="multiplicative", sp=reg_period)),
    ("damped", ExponentialSmoothing(trend="add", damped=True, seasonal="multiplicative", sp=reg_period))
])
forecaster.fit(y_train)
y_pred = forecaster.predict(fh)
plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"]);
smape_loss(y_test, y_pred)
```

```python
from sktime.forecasting.model_selection import ForecastingGridSearchCV, SlidingWindowSplitter

forecaster = ReducedRegressionForecaster(regressor=regressor, window_length=reg_period, strategy="recursive")
param_grid = {"window_length": [7, 14, 28]}

# we fit the forecaster on the initial window, and then use temporal cross-validation to find the optimal parameter
cv = SlidingWindowSplitter(initial_window=int(len(y_train) * 0.5))
gscv = ForecastingGridSearchCV(forecaster, cv=cv, param_grid=param_grid)
gscv.fit(y_train)
y_pred = gscv.predict(fh)
plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"]);
smape_loss(y_test, y_pred)
```

```python
from sktime.forecasting.trend import PolynomialTrendForecaster
from sktime.transformers.single_series.detrend import Detrender

# liner detrending
forecaster = PolynomialTrendForecaster(degree=1)
transformer = Detrender(forecaster=forecaster)
yt = transformer.fit_transform(y_train)

# internally, the Detrender uses the in-sample predictions of the PolynomialTrendForecaster
forecaster = PolynomialTrendForecaster(degree=1)
fh_ins = -np.arange(len(y_train)) # in-sample forecasting horizon
y_pred = forecaster.fit(y_train).predict(fh=fh_ins)

plot_ys(y_train, y_pred, yt, labels=["y_train", "Fitted linear trend", "Residuals"])
```

```python
from sktime.forecasting.theta import ThetaForecaster
forecaster = ThetaForecaster(sp=reg_period)
forecaster.fit(y_train)
alpha = 0.15  # 85% prediction intervals
y_pred, pred_ints = forecaster.predict(fh, return_pred_int=True, alpha=alpha)
smape_loss(y_test, y_pred)

fig, ax = plot_ys(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"])
ax.fill_between(y_pred.index, pred_ints["lower"], pred_ints["upper"], alpha=0.1, color="green", label=f"{1 - alpha}% prediction intervals")
plt.legend();
```
